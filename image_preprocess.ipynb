# Resize

```python
import os
import numpy as np
import cv2
from PIL import Image

original_path = '/gdrive/MyDrive/image_data/Validation/sad/cropped_Resized_sad_AREA/'
resized_path = '/gdrive/MyDrive/image_data/Validation/sad/Resized_cropped_Resized_sad_AREA/'

file_list = os.listdir(original_path)
img_list = []

for item in file_list :
    if item.find('.jpg') is not -1 :
        img_list.append(item)

total_image = len(img_list)
index = 0

for name in img_list :

    img = Image.open('%s%s'%(original_path, name))
    img_array = np.array(img)
    img_resize = cv2.resize(img_array, (120,120), interpolation = cv2.INTER_AREA)
    img = Image.fromarray(img_resize)
    img.save('%s%s'%(resized_path, name))

    print(name + '   ' + str(index) + '/' + str(total_image))
    index = index + 1

# original AREA 197장 29초
```

# Crop

```python
import dlib
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt


def detect_faces(image):

    # Create a face detector
    face_detector = dlib.get_frontal_face_detector()

    # Run detector and get bounding boxes of the faces on image.
    detected_faces = face_detector(image, 1)
    face_frames = [(x.left(), x.top(),
                    x.right(), x.bottom()) for x in detected_faces]

    return face_frames

# Load image
original_path = '/gdrive/MyDrive/image_data/Validation/sad/Resized_sad_AREA/'
cropped_path = '/gdrive/MyDrive/image_data/Validation/sad/cropped_Resized_sad_AREA/'

file_list = os.listdir(original_path)
img_list = []

for item in file_list :
    if item.find('.jpg') is not -1 :
        img_list.append(item)

total_image = len(img_list)
index = 0

for name in img_list :

    image = io.imread('%s%s'%(original_path, name))
    detected_faces = detect_faces(image)
    for n, face_rect in enumerate(detected_faces):
        face = Image.fromarray(image).crop(face_rect)
    face.save('%s%s'%(cropped_path, name))

    print(name + '   ' + str(index) + '/' + str(total_image))
    index = index + 1

# GPU 사용 안함 216장 19분 8초
# GPU 사용 안함 360 * 270 Resize 후 crop 1분 53초
```

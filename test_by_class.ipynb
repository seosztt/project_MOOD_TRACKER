```python
anger_test_dir = 'anger_test'
sad_test_dir = 'sad_test'
neutrality_test_dir = 'neutrality_test'
embarrassment_test_dir = 'embarrassment_test'
happy_test_dir = 'happy_test'
unrest_test_dir = 'unrest_test'
hurt_test_dir = 'hurt_test'
```

# test by class. 500장씩

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# hurt

hurt_test_generator = test_datagen.flow_from_directory(
                 hurt_test_dir,
                 target_size = (413, 557),
                 batch_size = 32,
                 class_mode = 'categorical')

loss, accuracy = model.evaluate(hurt_test_generator,
                                steps = 16)

print('Loss = {:.5f}'.format(loss))
print('Accuracy = {:.5f}'.format(accuracy))
```
